```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
df = pd.read_excel('172.20.10.2.xls')
```

```python
stuff = df.to_numpy()
t,ax,ay,az = stuff.T
```

```python
#plot with matplotlib in separated figures and colors



from turtle import color


plt.figure(1)
plt.plot(t,ax)
plt.title('X acceleration vs Time')
plt.xlabel('Time (s)')
plt.ylabel('X acceleration (m/s^2)')
plt.grid(True)
plt.show()

plt.figure(2)
plt.plot(t,ay, color= 'yellow')
plt.title('Y acceleration vs Time')
plt.xlabel('Time (s)')
plt.ylabel('Y acceleration (m/s^2)')
plt.grid(True)
plt.show()

plt.figure(3)
plt.plot(t,az, color='red')
plt.title('Z acceleration vs Time')
plt.xlabel('Time (s)')
plt.ylabel('Z acceleration (m/s^2)')
plt.grid(True)
plt.show()

```

```python

dt = t[1]-t[0]
vx = [0]
vy = [0]
vz = [0]

for i in np.arange(len(t)-1):
    vx = vx + [vx[-1] + ax[i]*dt]
    vy = vy + [vy[-1] + ay[i]*dt]
    vz = vz + [vz[-1] + (az[i]-9.8)*dt]

plt.plot(t,vx, color='green', label= 'vx')
plt.plot(t,vy, color='yellow', label= 'vy')
plt.plot(t,vz, color='red', label= 'vz')
plt.title('velocity vs Time')
plt.xlabel('Time (s)')
plt.ylabel('velocity (m/s)')
plt.grid(True)
plt.legend()
plt.show()
```

```python

dt = t[1]-t[0]
x = [0]
y = [0]
z = [0]
for i in np.arange(len(t)-1):
    x = x + [x[-1] + vx[i]*dt]
    y = y + [y[-1] + vy[i]*dt]
    z = z + [z[-1] + (vz[i])*dt]

plt.plot(t,x, color='green', label= 'x')
plt.plot(t,y, color='yellow', label= 'y')
plt.plot(t,z, color='red', label= 'z')
plt.title('r [m] vs Time')
plt.xlabel('Time (s)')
plt.ylabel('r [m')
plt.grid(True)
plt.legend()
plt.show()
```
